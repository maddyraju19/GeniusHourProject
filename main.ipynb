```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller


ratings = [pd.read_csv("/workspaces/GeniusHourProject/Aaron Rodgers QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/Brett Favre QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/Dan Marino QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/Joe Montana QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/John Elway QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/Kurt Warner QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/Steve Young QB Data.csv"),
    pd.read_csv("/workspaces/GeniusHourProject/Tom Brady QB Data.csv"), 
    pd.read_csv("/workspaces/GeniusHourProject/Patrick Mahomes QB Data.csv")]
for rating in ratings:
    qbrate = list(rating["Rate"])
    for i in range(len(qbrate)):
        try:
            qbrate[i] = float(qbrate[i])
        except ValueError:
            qbrate[i] = None
    qbrate = pd.Series(qbrate).interpolate()  
    diff = qbrate.diff().diff().diff().dropna()
    print(adfuller(diff))
    plot_acf(diff, lags=len(diff)/2)
    plt.show()
    plot_pacf(diff, lags=len(diff)/2)
    plt.show()


'''
model = ARIMA(diff, order=(1, 0, 1))
model_fit = model.fit()
model_summary = model_fit.summary()
print(model_summary)
'''


```
